```python
# Many ML algorithms (Logistic Regression, SVM, KNN) are sensitive to extreme values.
```

```python
# IQR capping for 'age' and 'campaign'
for col in ['age', 'campaign']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)

print("✅ IQR capping applied to: 'age' and 'campaign'")

# Handle 'pdays' special code
df['pdays_flag'] = df['pdays'].apply(lambda x: 0 if x == 999 else 1)
df['pdays'] = df['pdays'].replace(999, -1)
print("✅ 'pdays' handled (flag + replaced 999 with -1)")

# Optional capping for 'previous'
Q1 = df['previous'].quantile(0.25)
Q3 = df['previous'].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR
df['previous'] = df['previous'].clip(upper=upper)
print("⚖️ 'previous' capped optionally at upper IQR bound")
```
